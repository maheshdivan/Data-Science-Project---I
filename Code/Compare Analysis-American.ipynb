```python
#Dependencies 
%matplotlib notebook
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import sem
from scipy.stats import linregress
import warnings 
import json
warnings.filterwarnings('ignore')
```

```python
# Read in csv
csv_path_1="lga_american_date"
csv_path_2="lga_american_dest"
csv_path_3="JFK_AA_date.csv"
csv_path_4="jfk_aa_dest.csv"
csv_path_5="EWR_AA_date.csv"
csv_path_6="EWR_AA_delay_des_reason.csv"


american_lga_date=pd.read_csv(csv_path_1)
american_lga_dest=pd.read_csv(csv_path_2)
american_jfk_date=pd.read_csv(csv_path_3)
american_jfk_dest=pd.read_csv(csv_path_4)
american_ewr_date=pd.read_csv(csv_path_5)
american_ewr_dest=pd.read_csv(csv_path_6)


american_jfk_date.head()
```

```python
american_lga_date['Airport Name']='LGA'
american_lga_dest['Airport Name']='LGA'
american_jfk_date['Airport Name']='JFK'
american_jfk_dest['Airport Name']='JFK'
american_ewr_date['Airport Name']='EWR'
american_ewr_dest['Airport Name']='EWR'
```

```python
american_jfk_date.head()
```

```python
final_compare_date=pd.concat([american_lga_date, american_jfk_date, american_ewr_date])
```

```python
final_compare_date.head()
```

```python
final_compare_date.columns
```

```python
final_compare_date_1=final_compare_date.drop(['Actual elapsed time (Minutes)'], axis=1)
```

```python
final_compare_date_1
```

```python
final_compare_date_2=final_compare_date_1.drop(['Date (MM/DD/YYYY)'], axis=1)
final_compare_date_2
```

```python
final_compare_date_2=final_compare_date_2.groupby(['Airport Name'], as_index=False).sum()
final_compare_date_2
```

```python
final_compare_date_3=final_compare_date_2.drop(['Scheduled elapsed time (Minutes)'], axis=1)
final_compare_date_3
```

```python
final_compare_date_sort=final_compare_date_3.sort_values(by=['Departure delay (Minutes)'], ascending=False)
final_compare_date_sort.count()
```
